```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
import warnings 

warnings.filterwarnings('ignore')
```

```python
''' reading dataset '''
df1 = pd.read_csv('2018-19_AutonomousVehicleDisengagementReports(firsttimefilers).csv')
df2 = pd.read_csv('2019AutonomousVehicleDisengagementReports.csv')
```

```python
''' concatenating df1 and df2 '''
df = pd.concat([df1, df2])

df.head()
```

```python
''' renaming columns '''
#Renaming Columns for Legibility 
df.columns = ['Manufacturer', 'Permit Number', 'Date', 'VIN number','Vehicle Is Capable Of Operating Without A Driver',
              'Driver Present','Disengagement Initiated By','Disengagement Location', 
              'Description Of Facts Causing Disengagement', 'N/A','N/A'] 
```

```python
df.head()
```

```python
df["Disengagement Location"].unique()
```

```python
''' cleaning dataset '''
def clean_data(x):
    if x == "street (high speed)" or x == "STREET" or x == "Street" or x == "Downtown street" or x == "downtown street":
        x = 'street'
        
    elif x ==  "Highway":
        x = 'highway'
        
    elif x == "Freeway":
        x = 'freeway'
        
    elif x == "Rural":
        x = 'rural road'
    
    elif x == "Parking Facility":
        x = "parking facility"
    
    elif x == "Interstate":
        x = "interstate"
    
    return str(x).lower()

df["Disengagement Location"] = df["Disengagement Location"].map(clean_data)
```

```python
''' converting every word into lower case in driver present column '''
df["Driver Present"] = df["Driver Present"].str.lower()

''' converting every word into lower case in Vehicle Is Capable Of Operating Without A Driver column '''
df["Vehicle Is Capable Of Operating Without A Driver"] = df["Vehicle Is Capable Of Operating Without A Driver"].str.lower()
```

```python
def clean_data(x):
    if x == "Test Driver" or x == "Safety Driver" or x == "Vehicle Operator" or x == "test driver":
        x = "test driver"
    return str(x).lower()
    
df["Disengagement Initiated By"] = df["Disengagement Initiated By"].map(clean_data)
```

```python
''' changing date '''
df["Date"] = df["Date"].replace("1/30.2019", "1/30/2019")

''' converting date column into date time '''
df["Date"] = pd.to_datetime(df["Date"]) 

''' set date column as index'''
df = df.set_index("Date") 
```

```python
''' barplot of Disengage Location '''
plt.figure(figsize=(10, 5))
plt.xticks(rotation=90)
sns.barplot(df["Disengagement Location"].value_counts().index, df['Disengagement Location'].value_counts());
plt.xlabel('Disengagement Location ', fontsize=15);
plt.ylabel('Count', fontsize=15);
```
